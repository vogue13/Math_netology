Уровень 0:
Даны вектора x=(1,1), y=(2, 0) и z = (0,2)  

Надо:
Изобразить вектора на экране  
Изобразить точку x + y + z  
Найти угол между векторами x и y



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x =  np.array([1,1])
y =  np.array([2,0])
z = np.array([0,2])
```

```python
plt.plot([0,x[0]], [0,x[1]])
plt.plot([0,y[0]], [0,y[1]])
plt.plot([0,z[0]], [0,z[1]])
plt.show()
```

```python
w = x + y + z
w
```

```python
plt.plot(w[0], w[1], c='b', marker='o')
plt.plot([0,x[0]], [0,x[1]])
plt.plot([0,y[0]], [0,y[1]])
plt.plot([0,z[0]], [0,z[1]])

plt.show()
```

```python
cosA = np.dot(x,y) / ( (np.dot(x,x)**0.5) * (np.dot(y,y)**0.5) )
cosA
```
